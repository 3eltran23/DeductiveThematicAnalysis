## Agreement: Notation and Definitions

To the reliability study,the agreement procedure suggests itselt is that of having two or more judges independtly categorize as sample of units. For this reason,  we need to define the following notation:



*   The unlabeled datasets of $n$ items, $I =\{i_1, i_2, i_i, \dots, i_n\}$
*   The sampling of $s$ item of the set $I$, $S \sim U(I)$, where $s \leq n$
*   The set of $k$ categories, $K= \{c_1, c_2, \dots, c_k \}$
*   The set of judges, $J = \{j_1, \dots , j_m\}$, where $m \geq 2$
*   The item label set, $\mathbf{y^{j_i}} = [ y_{1}^{j_i}, \dots,  y_{n}^{j_i} ]^T$ is a $n$ dimensional vector with a category label $y_i \in K$ assigned to the item i for the judge $j_i$

Finally, we use $q$ with a subscript to indicate the number of judgments of a given type:


*   $q_{ik}$ is the number of coders who assigned item $i$ to category $k$
*   $q_{j_c,k}$ is the number of items assigned by coder $j_c$ to category $k$
*   $q_k$ is the number of items assigned by all coders to category $k$


* $p_{j_c,k}$, the proportion $p$ of items assigne by coder $j_c$ to category $k$ is given by  $q_{j_c,k} / n$ where n is the number of items.

To compute the inter-coder agreement, the following conditions can be taken as assumptions:



1.   The units or items $i_i \in I$ are independent.
2.   The categories $c_i \in K$ of the nominal scale are independent, mutually exclusive, and exhaustive.
3. The judges operate independently.
4. There is no restriction placed on the distribution of judgments over categories for either judge.

It is important to note that there is no criterion for the correctness of judgements, therefore we supponse that judges are equally competent to make judgements.












## Inter-Coder Agreement Approaches

### Inter-Coder Agreement: Naive Approach
The most primitive measure of agreement between two judges ($J=\{j_1, j_2\}$) is counting the proportion of cases in which the judges agreed (a.k.a observer agreement). The observed agreement is defiend as the percentage of judgemets on which the two judges agree when coding the same data independtly.  This is the number of items on which the coder agree divided by the total number of items.

The agreement value $agr_i$ for the item $i$ is defined as follow
$$
agr_i = \begin{cases}
1 & \text{if  }  y_i^{J_1} = y_i^{J_2},\\
0 & \text{otherwise}
\end{cases}
$$

Observer agreement over the values $agr_i$ for all the n items in $S$ is then:

$$
A_0 = \frac{1}{n} \sum_{i=1}^{n} agr_i
$$

#### Example
Consider a sample of 200 items (i.e., $n=200$) in which the Judge $j_1$ and Judge $j_2$ indepently categorize as follows:

|         |             | $j_1$ |       |       |            |
|---------|-------------|-------|-------|-------|------------|
|         | $K$         | $c_1$ | $c_2$ | $c_3$ | $q_{j_2,Ck}$|
|         |-------------|-------|-------|-------|------------|
|         |  $c_1$      | **50**|  26   |  24   |    100     |
| $j_2$   |  $c_2$      | 24    |  **4**|  32   |     60     |
|         |  $c_3$      | 6     |  30   |  **4**|     40     |
|         |  $q_{j_1,ck}$| 80    | 60    | 60    | 200        |

In the table, bold font indicate the number of items in which the judges agreed for each category. Therefore, $A_0$ from the table is obtained by summing the cell on the diagonal and dividing by the total numer of items (i.e., n=200).


$$
A_0 = \frac{50 + 4 + 4}{200} = 0.29
$$

Therefore, for above Table, there is a $.29 $ observered agreement. It is important to note this approach exhibit the following disadvantages:



*  The percentage agreement $A_o$ is biased in favor of dimensions with a
small number of categories. In other words, given two coding schemes for the same phenomemn, the one with fewer categories will result in higher percentage just by chance Scott (1955).   

*  The Percentage agreement cannot be trusted is that it does not correct for the distribution of items among categories: We expect a higher percentage agreement when one category is much more common than the other.


In the following, we will discuss another approaches to try to fixed this approach just by considered the expected agreement.

### Expected Agreement vs Observed Agreement: $\chi^2$ Chi-Square

Once you have the matrix of proportions of agreement, you can apply the chi-square test to examine whether there is a statistically significant difference between the observed proportions and the proportions expected under the null hypothesis of independence. It is a non-parametric test, meaning it does not assume any underlying data distribution.

For applying $\chi^2$ Test, the data must meet the following requirements:

* Two categorical variables (judges)
* Relatively large sample size
* Categories of variables (two or more)
* Independence of observations

#### Steps for Conducting the  $\chi^2$  Test



1.   **Formulate Hypotheses**: State the null and alternative hypotheses. The null hypotesis assert that the proportions of observerd agreement, between $j_1$ and $j_2$, is equal to the proportions of the agreementet expected. $A_o = A_e$. While the alternative hypotesis suggest otherwise.

2.   **Calcualte Expected Proportions**. Compute the expected proportions under the assumption of independence. This is often done by multiplying the row and column totals to get the expected cell frequencies and then converting them to proportions.

3. **Compute the $\chi^2$ (Chi-Square) Statistic**: Compute the $\chi^2$ statistic using the formula:

$$
\chi^2 = \sum \frac{(O_i-E_i)^2}{E_i}
$$

where $O_i$ is the observed proportion, $E_i$ is the expected, and the sum is taken over all cell in the contigency table.

**Observed proportions**

|---------|-------------|-------|-------|-------|------------|

|         |  $c_1$      | **0.25**|  0.13   |  0.12   |    0.50 |

Therefore, Cohen (1960) present a coefficient to measure the degree of agreement in nominal scales, and to provide means of testing hypotheses and setting confidence limits for this coefficient.






```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import chi2_contingency

# Define the data for the rows
observed = np.array([[50, 26, 24],
                     [24, 4, 32],
                     [6, 30, 4]])
# Compute chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

#df = 4
alpha = 0.05
critical_value = stats.chi2.ppf(1-alpha, df)
calculated_chi_square = chi2

# Generate values for the chi-square distribution
x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df)

plt.plot(x, y, label=f'Chi-Square Distribution (df={df})')
plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.5, label='Critical Region')
plt.axvline(calculated_chi_square, color='blue', linestyle='dashed', label='Calculated Chi-Square')
plt.axvline(critical_value, color='green', linestyle='dashed', label='Critical Value')
plt.title('Chi-Square Distribution and Critical Region')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()


print("Chi-square statistic:", calculated_chi_square)
print("p-value:", p)
print("significance level: ", alpha)
print("Critical value:", critical_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)


# Check for significance
if p < alpha:
    print("The result is significant at the", alpha, "level. Reject the null hypothesis.")
else:
    print("The result is not significant at the", alpha, "level. Fail to reject the null hypothesis.")
```

### Inter-Coder Agreement: Cofficient of Agreement

For any problem in nominal scale agreement between two judges, are only two relevant quantities:

$$
A_o = \text{the proportion of units in which the judges agreed} \\
A_e = \text{the proportion of units for which agreement is expected by chance}
$$

The coefficent $\kappa$ is the proportion of agreement after chance agreement is removed from consideration:

$$
k = \frac{A_o - A_e}{1- A_e}
$$

where  $1-A_e$ measure how much agreement over and above chance is attainable. $A_o - A_e$, represents the proportions of cases in which beyond-chance agreement ocurred and is the number of the coefficent.


#### Range of $\kappa$ Values.

The range of Cohen's kappa can vary from -1 to 1.



$$
\kappa =
\begin{cases}
1 & \text{it indicates perfect agreement }\\
0 & \text{it suggest agreement equivalence to chance} A_o = A_e \\
< 0 & \text{it implies agreement worse than would be expected by chance}  A_o < A_e \\
\end{cases}
$$


For our example the  Cohen's kappa is as follow:

$$
k = \frac{.29 - .35}{1- .35} = \frac{-.06}{0.65} = -0.09
$$
